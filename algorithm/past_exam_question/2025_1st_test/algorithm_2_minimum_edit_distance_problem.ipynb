## ğŸ’¡ ì ‘ê·¼ ë°©ë²•

### ğŸ”¹ í•µì‹¬ ì•„ì´ë””ì–´: Dynamic Programming (DP)

* **DP ì •ì˜:**
  `dp[i][j]` = `word1[:i]` â†’ `word2[:j]` ë¡œ ë³€í™˜í•˜ëŠ” ìµœì†Œ í¸ì§‘ íšŸìˆ˜

* **ì í™”ì‹:**

  ```
  if word1[i-1] == word2[j-1]:
      dp[i][j] = dp[i-1][j-1]
  else:
      dp[i][j] = min(
          dp[i-1][j] + 1,      # ì‚­ì œ
          dp[i][j-1] + 1,      # ì‚½ì…
          dp[i-1][j-1] + 1     # ì¹˜í™˜
      )
  ```

* **ì´ˆê¸°ê°’:**

  * `dp[i][0] = i` â†’ ië²ˆ ì‚­ì œ
  * `dp[0][j] = j` â†’ jë²ˆ ì‚½ì…


````python
# ============================================================
# ë¬¸ì œ: Edit Distance (ìµœì†Œ í¸ì§‘ ê±°ë¦¬, Levenshtein Distance)
# ì•Œê³ ë¦¬ì¦˜: Dynamic Programming (ë™ì  í”„ë¡œê·¸ë˜ë°)
# ì‹œê°„ ë³µì¡ë„: O(m Ã— n) - m, nì€ ê° ë¬¸ìì—´ì˜ ê¸¸ì´
# ê³µê°„ ë³µì¡ë„: O(m Ã— n) - DP í…Œì´ë¸” í¬ê¸° (ìµœì í™” ì‹œ O(min(m,n)))
# ============================================================
#
# ğŸ“Œ ë¬¸ì œ í•µì‹¬:
# - ë¬¸ìì—´ word1ì„ word2ë¡œ ë³€í™˜í•˜ëŠ” ìµœì†Œ ì—°ì‚° íšŸìˆ˜
# - í—ˆìš© ì—°ì‚°: ì‚½ì…(Insert), ì‚­ì œ(Delete), ì¹˜í™˜(Replace)
#
# ğŸ“Œ ì‹¤ìƒí™œ ì‘ìš©:
# - ë§ì¶¤ë²• ê²€ì‚¬ (ì˜¤íƒ€ êµì •)
# - DNA ì„œì—´ ë¹„êµ (ìƒë¬¼ì •ë³´í•™)
# - ìì—°ì–´ ì²˜ë¦¬ (ë¬¸ì¥ ìœ ì‚¬ë„)
# - Git diff (íŒŒì¼ ë³€ê²½ ì‚¬í•­ ì¶”ì )
#
# ğŸ“Œ DP ì •ì˜:
# dp[i][j] = word1ì˜ ì²˜ìŒ iê°œ ë¬¸ìë¥¼ word2ì˜ ì²˜ìŒ jê°œ ë¬¸ìë¡œ 
#            ë³€í™˜í•˜ëŠ” ë° í•„ìš”í•œ ìµœì†Œ í¸ì§‘ íšŸìˆ˜
#
# ğŸ“Œ ì í™”ì‹ (Recurrence Relation):
# if word1[i-1] == word2[j-1]:
#     dp[i][j] = dp[i-1][j-1]  # ë¬¸ìê°€ ê°™ìœ¼ë©´ ì—°ì‚° ë¶ˆí•„ìš”
# else:
#     dp[i][j] = min(
#         dp[i-1][j] + 1,      # ì‚­ì œ: word1ì—ì„œ ë¬¸ì ì œê±°
#         dp[i][j-1] + 1,      # ì‚½ì…: word1ì— ë¬¸ì ì¶”ê°€
#         dp[i-1][j-1] + 1     # ì¹˜í™˜: word1ì˜ ë¬¸ìë¥¼ word2ì˜ ë¬¸ìë¡œ ë³€ê²½
#     )
# ============================================================

class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        """
        ë‘ ë¬¸ìì—´ ì‚¬ì´ì˜ ìµœì†Œ í¸ì§‘ ê±°ë¦¬ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
        
        Args:
            word1: ì›ë³¸ ë¬¸ìì—´
            word2: ëª©í‘œ ë¬¸ìì—´
            
        Returns:
            int: word1ì„ word2ë¡œ ë³€í™˜í•˜ëŠ” ìµœì†Œ ì—°ì‚° íšŸìˆ˜
            
        Examples:
            >>> sol = Solution()
            >>> sol.minDistance("kitten", "sitting")
            3  # kitten â†’ sitten â†’ sittin â†’ sitting
        """
        
        m, n = len(word1), len(word2)
        
        # ============================================================
        # ì—£ì§€ ì¼€ì´ìŠ¤ ì²˜ë¦¬
        # ============================================================
        # word1ì´ ë¹„ì–´ìˆìœ¼ë©´ word2 ê¸¸ì´ë§Œí¼ ì‚½ì… í•„ìš”
        if m == 0:
            return n
        # word2ê°€ ë¹„ì–´ìˆìœ¼ë©´ word1 ê¸¸ì´ë§Œí¼ ì‚­ì œ í•„ìš”
        if n == 0:
            return m
        
        # ============================================================
        # DP í…Œì´ë¸” ì´ˆê¸°í™”
        # ============================================================
        # dp[i][j]: word1[:i]ë¥¼ word2[:j]ë¡œ ë³€í™˜í•˜ëŠ” ìµœì†Œ ë¹„ìš©
        # í…Œì´ë¸” í¬ê¸°: (m+1) x (n+1)
        # +1ì€ ë¹ˆ ë¬¸ìì—´("")ë„ í¬í•¨í•˜ê¸° ìœ„í•¨
        dp = [[0] * (n + 1) for _ in range(m + 1)]
        
        # ============================================================
        # ë² ì´ìŠ¤ ì¼€ì´ìŠ¤ (Base Case) ì„¤ì •
        # ============================================================
        # dp[i][0]: word1ì˜ iê°œ ë¬¸ìë¥¼ ë¹ˆ ë¬¸ìì—´ë¡œ ë³€í™˜ â†’ ië²ˆ ì‚­ì œ
        for i in range(m + 1):
            dp[i][0] = i
        
        # dp[0][j]: ë¹ˆ ë¬¸ìì—´ì„ word2ì˜ jê°œ ë¬¸ìë¡œ ë³€í™˜ â†’ jë²ˆ ì‚½ì…
        for j in range(n + 1):
            dp[0][j] = j
        
        # ============================================================
        # DP í…Œì´ë¸” ì±„ìš°ê¸° (Bottom-Up)
        # ============================================================
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                # í˜„ì¬ ë¹„êµí•˜ëŠ” ë¬¸ì (ì¸ë±ìŠ¤ ì£¼ì˜: dpëŠ” 1-based, ë¬¸ìì—´ì€ 0-based)
                char1 = word1[i - 1]
                char2 = word2[j - 1]
                
                if char1 == char2:
                    # ë¬¸ìê°€ ê°™ìœ¼ë©´ ì¶”ê°€ ì—°ì‚° ë¶ˆí•„ìš”
                    # ì´ì „ ìƒíƒœ(dp[i-1][j-1])ë¥¼ ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜´
                    dp[i][j] = dp[i - 1][j - 1]
                else:
                    # ë¬¸ìê°€ ë‹¤ë¥´ë©´ ì„¸ ê°€ì§€ ì—°ì‚° ì¤‘ ìµœì†Œ ì„ íƒ
                    dp[i][j] = min(
                        dp[i - 1][j] + 1,      # ì‚­ì œ: word1ì˜ ë§ˆì§€ë§‰ ë¬¸ì ì œê±°
                        dp[i][j - 1] + 1,      # ì‚½ì…: word1ì— word2ì˜ ë¬¸ì ì¶”ê°€
                        dp[i - 1][j - 1] + 1   # ì¹˜í™˜: word1ì˜ ë¬¸ìë¥¼ word2ì˜ ë¬¸ìë¡œ ë³€ê²½
                    )
        
        # ìµœì¢… ê²°ê³¼: word1 ì „ì²´ë¥¼ word2 ì „ì²´ë¡œ ë³€í™˜í•˜ëŠ” ë¹„ìš©
        return dp[m][n]
    
    
    def minDistance_optimized(self, word1: str, word2: str) -> int:
        """
        ê³µê°„ ìµœì í™” ë²„ì „ - O(min(m,n)) ê³µê°„ ì‚¬ìš©
        
        DP í…Œì´ë¸”ì—ì„œ ê° í–‰ì€ ì´ì „ í–‰ë§Œ ì°¸ì¡°í•˜ë¯€ë¡œ,
        ì „ì²´ í…Œì´ë¸” ëŒ€ì‹  ë‘ ê°œì˜ í–‰ë§Œ ìœ ì§€í•˜ë©´ ë©ë‹ˆë‹¤.
        
        ì¶”ê°€ ìµœì í™”: ì§§ì€ ë¬¸ìì—´ì„ word2ë¡œ ì‚¬ìš©í•˜ì—¬ ê³µê°„ ì ˆì•½
        """
        
        # ì§§ì€ ë¬¸ìì—´ì„ word2ë¡œ ì„¤ì • (ê³µê°„ ìµœì í™”)
        if len(word1) < len(word2):
            word1, word2 = word2, word1
        
        m, n = len(word1), len(word2)
        
        if n == 0:
            return m
        
        # ì´ì „ í–‰ê³¼ í˜„ì¬ í–‰ë§Œ ìœ ì§€
        prev_row = list(range(n + 1))  # ë² ì´ìŠ¤ ì¼€ì´ìŠ¤: [0, 1, 2, ..., n]
        curr_row = [0] * (n + 1)
        
        for i in range(1, m + 1):
            curr_row[0] = i  # ë² ì´ìŠ¤ ì¼€ì´ìŠ¤: ië²ˆ ì‚­ì œ
            
            for j in range(1, n + 1):
                if word1[i - 1] == word2[j - 1]:
                    curr_row[j] = prev_row[j - 1]
                else:
                    curr_row[j] = min(
                        prev_row[j] + 1,      # ì‚­ì œ
                        curr_row[j - 1] + 1,  # ì‚½ì…
                        prev_row[j - 1] + 1   # ì¹˜í™˜
                    )
            
            # í–‰ êµì²´ (í˜„ì¬ í–‰ì´ ë‹¤ìŒ ë°˜ë³µì˜ ì´ì „ í–‰ì´ ë¨)
            prev_row, curr_row = curr_row, prev_row
        
        return prev_row[n]
    
    
    def minDistance_with_path(self, word1: str, word2: str) -> tuple:
        """
        ìµœì†Œ í¸ì§‘ ê±°ë¦¬ì™€ í•¨ê»˜ ì‹¤ì œ ë³€í™˜ ê³¼ì •ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
        
        Returns:
            tuple: (ìµœì†Œ ê±°ë¦¬, ë³€í™˜ ê³¼ì • ë¦¬ìŠ¤íŠ¸)
        """
        
        m, n = len(word1), len(word2)
        
        # DP í…Œì´ë¸”
        dp = [[0] * (n + 1) for _ in range(m + 1)]
        
        # ì—°ì‚° ì¶”ì  í…Œì´ë¸”: ì–´ë–¤ ì—°ì‚°ì„ ì„ íƒí–ˆëŠ”ì§€ ê¸°ë¡
        # 0: ë™ì¼, 1: ì‚­ì œ, 2: ì‚½ì…, 3: ì¹˜í™˜
        ops = [[0] * (n + 1) for _ in range(m + 1)]
        
        # ë² ì´ìŠ¤ ì¼€ì´ìŠ¤
        for i in range(m + 1):
            dp[i][0] = i
            ops[i][0] = 1  # ì‚­ì œ
        for j in range(n + 1):
            dp[0][j] = j
            ops[0][j] = 2  # ì‚½ì…
        ops[0][0] = 0
        
        # DP ì±„ìš°ê¸°
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if word1[i - 1] == word2[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]
                    ops[i][j] = 0  # ë™ì¼
                else:
                    delete = dp[i - 1][j] + 1
                    insert = dp[i][j - 1] + 1
                    replace = dp[i - 1][j - 1] + 1
                    
                    dp[i][j] = min(delete, insert, replace)
                    
                    if dp[i][j] == delete:
                        ops[i][j] = 1
                    elif dp[i][j] == insert:
                        ops[i][j] = 2
                    else:
                        ops[i][j] = 3
        
        # ì—­ì¶”ì ìœ¼ë¡œ ë³€í™˜ ê³¼ì • ë³µì›
        path = []
        i, j = m, n
        
        while i > 0 or j > 0:
            if ops[i][j] == 0:  # ë™ì¼
                path.append(f"'{word1[i-1]}' == '{word2[j-1]}' (ìœ ì§€)")
                i -= 1
                j -= 1
            elif ops[i][j] == 1:  # ì‚­ì œ
                path.append(f"'{word1[i-1]}' ì‚­ì œ (Delete)")
                i -= 1
            elif ops[i][j] == 2:  # ì‚½ì…
                path.append(f"'{word2[j-1]}' ì‚½ì… (Insert)")
                j -= 1
            else:  # ì¹˜í™˜
                path.append(f"'{word1[i-1]}' â†’ '{word2[j-1]}' (Replace)")
                i -= 1
                j -= 1
        
        path.reverse()
        return dp[m][n], path


# ============================================================
# ì•Œê³ ë¦¬ì¦˜ ì‹œê°ì  ì„¤ëª…
# ============================================================
def visualize_dp_table(word1: str, word2: str):
    """
    DP í…Œì´ë¸”ì„ ì‹œê°ì ìœ¼ë¡œ ì¶œë ¥í•©ë‹ˆë‹¤.
    """
    m, n = len(word1), len(word2)
    
    # DP í…Œì´ë¸” ìƒì„±
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1
    
    # í…Œì´ë¸” ì¶œë ¥
    print(f"\n{'='*60}")
    print(f"Edit Distance: '{word1}' â†’ '{word2}'")
    print(f"{'='*60}")
    
    # í—¤ë”
    header = "     " + "  " + "  ".join(f" {c}" for c in " " + word2)
    print(header)
    print("    " + "-" * (len(header) - 4))
    
    # DP í…Œì´ë¸” í–‰
    for i in range(m + 1):
        row_char = " " if i == 0 else word1[i - 1]
        row = f" {row_char} |"
        for j in range(n + 1):
            row += f" {dp[i][j]:2d}"
        print(row)
    
    print(f"\nìµœì†Œ í¸ì§‘ ê±°ë¦¬: {dp[m][n]}")
    
    # ë³€í™˜ ê³¼ì •
    sol = Solution()
    dist, path = sol.minDistance_with_path(word1, word2)
    print(f"\në³€í™˜ ê³¼ì •:")
    for step in path:
        print(f"  â€¢ {step}")


# ============================================================
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ ë° ìë™ ì±„ì 
# ============================================================
def test_minDistance():
    sol = Solution()
    test_cases = [
        ("kitten", "sitting", 3),
        ("abc", "yabd", 2),
        ("", "abc", 3),
        ("a", "a", 0),
        (" ", "a", 1),
        ("intention", "execution", 5),
        ("horse", "ros", 3),
        ("", "", 0),
        ("algorithm", "altruistic", 6),
    ]

    score = 0
    print("=" * 60)
    print("ìë™ ì±„ì  ê²°ê³¼")
    print("=" * 60)
    
    for idx, (A, B, expected) in enumerate(test_cases, 1):
        result = sol.minDistance(A, B)
        correct = (result == expected)
        print(f"Test {idx}: '{A}' â†’ '{B}' â†’ ê²°ê³¼={result}, ì˜ˆìƒ={expected} â†’ {'âœ… ì •ë‹µ' if correct else 'âŒ ì˜¤ë‹µ'}")
        if correct:
            score += 1

    print(f"\nì´ ì ìˆ˜: {score}/{len(test_cases)}")


# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
test_minDistance()

# DP í…Œì´ë¸” ì‹œê°í™” ì˜ˆì‹œ
visualize_dp_table("kitten", "sitting")
visualize_dp_table("abc", "yabd")
````
