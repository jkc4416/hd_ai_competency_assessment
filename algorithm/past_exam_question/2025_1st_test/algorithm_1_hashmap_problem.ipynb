````python
# ============================================================
# ë¬¸ì œ: Find First Unique Character (ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ì°¾ê¸°)
# ì•Œê³ ë¦¬ì¦˜: HashMap (í•´ì‹œë§µ) í™œìš©
# ì‹œê°„ ë³µì¡ë„: O(n) - ë¬¸ìì—´ì„ ë‘ ë²ˆ ìˆœíšŒ
# ê³µê°„ ë³µì¡ë„: O(1) - ì˜ë¬¸ ì†Œë¬¸ì 26ê°œë¡œ ì œí•œ
# ============================================================
#
# ğŸ“Œ ë¬¸ì œ í•µì‹¬:
# - ë¬¸ìì—´ì—ì„œ í•œ ë²ˆë§Œ ë“±ì¥í•˜ëŠ” ì²« ë²ˆì§¸ ë¬¸ìì˜ ì¸ë±ìŠ¤ ë°˜í™˜
# - ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µì´ë©´ -1 ë°˜í™˜
#
# ğŸ“Œ í•´ê²° ì „ëµ:
# 1. ì²« ë²ˆì§¸ ìˆœíšŒ: ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ë¥¼ í•´ì‹œë§µì— ì €ì¥
# 2. ë‘ ë²ˆì§¸ ìˆœíšŒ: ë“±ì¥ íšŸìˆ˜ê°€ 1ì¸ ì²« ë²ˆì§¸ ë¬¸ìì˜ ì¸ë±ìŠ¤ ë°˜í™˜
#
# ğŸ“Œ ì™œ í•´ì‹œë§µì„ ì‚¬ìš©í•˜ëŠ”ê°€?
# - í•´ì‹œë§µì˜ ì¡°íšŒ/ì‚½ì…ì€ O(1) ì‹œê°„ ë³µì¡ë„
# - ë¸Œë£¨íŠ¸í¬ìŠ¤(O(nÂ²))ë³´ë‹¤ í›¨ì”¬ íš¨ìœ¨ì 
# ============================================================

from collections import Counter

class Solution:
    def firstUniqChar(self, s: str) -> int:
        """
        ë¬¸ìì—´ì—ì„œ í•œ ë²ˆë§Œ ë“±ì¥í•˜ëŠ” ì²« ë²ˆì§¸ ë¬¸ìì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
        
        Args:
            s: ì…ë ¥ ë¬¸ìì—´ (ì†Œë¬¸ì ì˜ì–´ ì•ŒíŒŒë²³ìœ¼ë¡œë§Œ êµ¬ì„±)
            
        Returns:
            int: ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ìì˜ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ -1
            
        Examples:
            >>> sol = Solution()
            >>> sol.firstUniqChar("morningstar")
            0  # 'm'ì´ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì
            >>> sol.firstUniqChar("aabb")
            -1  # ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µ
        """
        
        # ============================================================
        # ë°©ë²• 1: collections.Counter ì‚¬ìš© (ê°€ì¥ ê°„ê²°)
        # ============================================================
        # CounterëŠ” í•´ì‹œë§µ ê¸°ë°˜ìœ¼ë¡œ ê° ì›ì†Œì˜ ê°œìˆ˜ë¥¼ ìë™ìœ¼ë¡œ ì„¸ì–´ì¤ë‹ˆë‹¤.
        # ì˜ˆ: Counter("aab") â†’ {'a': 2, 'b': 1}
        
        # ì—£ì§€ ì¼€ì´ìŠ¤: ë¹ˆ ë¬¸ìì—´
        if not s:
            return -1
        
        # Step 1: ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ë¥¼ ê³„ì‚°
        # CounterëŠ” ë‚´ë¶€ì ìœ¼ë¡œ ë”•ì…”ë„ˆë¦¬(í•´ì‹œë§µ)ë¥¼ ì‚¬ìš©
        char_count = Counter(s)
        
        # Step 2: ë¬¸ìì—´ì„ ìˆœíšŒí•˜ë©° ë“±ì¥ íšŸìˆ˜ê°€ 1ì¸ ì²« ë²ˆì§¸ ë¬¸ì ì°¾ê¸°
        for idx, char in enumerate(s):
            if char_count[char] == 1:
                return idx
        
        # ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µì¸ ê²½ìš°
        return -1
    
    
    def firstUniqChar_manual(self, s: str) -> int:
        """
        ë°©ë²• 2: ì¼ë°˜ ë”•ì…”ë„ˆë¦¬ ì‚¬ìš© (Counter ì—†ì´ ì§ì ‘ êµ¬í˜„)
        
        í•™ìŠµ ëª©ì ìœ¼ë¡œ Counter ì—†ì´ ì§ì ‘ í•´ì‹œë§µì„ êµ¬í˜„í•©ë‹ˆë‹¤.
        ì‹¤ì œ ë©´ì ‘ì—ì„œ Counter ì‚¬ìš©ì´ ì œí•œë  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤.
        """
        
        if not s:
            return -1
        
        # Step 1: ë¹ˆ í•´ì‹œë§µ(ë”•ì…”ë„ˆë¦¬) ìƒì„±
        char_count = {}
        
        # Step 2: ì²« ë²ˆì§¸ ìˆœíšŒ - ê° ë¬¸ìì˜ ë¹ˆë„ ê³„ì‚°
        # ë¬¸ìì—´ì˜ ê° ë¬¸ìë¥¼ ìˆœíšŒí•˜ë©´ì„œ ì¹´ìš´íŠ¸ ì¦ê°€
        for char in s:
            # í•´ì‹œë§µì— ë¬¸ìê°€ ì—†ìœ¼ë©´ 0ìœ¼ë¡œ ì´ˆê¸°í™” í›„ 1 ì¦ê°€
            # dict.get(key, default): keyê°€ ì—†ìœ¼ë©´ default ë°˜í™˜
            char_count[char] = char_count.get(char, 0) + 1
        
        # Step 3: ë‘ ë²ˆì§¸ ìˆœíšŒ - ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ì°¾ê¸°
        for idx, char in enumerate(s):
            if char_count[char] == 1:
                return idx
        
        return -1
    
    
    def firstUniqChar_optimized(self, s: str) -> int:
        """
        ë°©ë²• 3: ì¸ë±ìŠ¤ ì €ì¥ ë°©ì‹ (í•œ ë²ˆì˜ ìˆœíšŒë¡œ í•´ê²° ì‹œë„)
        
        ê° ë¬¸ìì˜ ì²« ë“±ì¥ ì¸ë±ìŠ¤ë¥¼ ì €ì¥í•˜ê³ ,
        ì¤‘ë³µ ë°œìƒ ì‹œ -1ë¡œ ë§ˆí‚¹í•˜ëŠ” ë°©ì‹ì…ë‹ˆë‹¤.
        
        ì¥ì : ë¬¸ìì—´ì´ ë§¤ìš° ê¸¸ê³  ìœ ì¼ ë¬¸ìê°€ ì•ìª½ì— ìˆìœ¼ë©´ ë” íš¨ìœ¨ì 
        ë‹¨ì : êµ¬í˜„ì´ ë³µì¡í•˜ê³ , ìµœì•…ì˜ ê²½ìš° ì„±ëŠ¥ ì°¨ì´ ì—†ìŒ
        """
        
        if not s:
            return -1
        
        # ê° ë¬¸ìì˜ ì²« ë“±ì¥ ì¸ë±ìŠ¤ ì €ì¥ (-1: ì¤‘ë³µ ë°œìƒ)
        first_index = {}
        
        for idx, char in enumerate(s):
            if char not in first_index:
                # ì²˜ìŒ ë“±ì¥: ì¸ë±ìŠ¤ ì €ì¥
                first_index[char] = idx
            else:
                # ì¤‘ë³µ ë°œìƒ: -1ë¡œ ë§ˆí‚¹ (ë” ì´ìƒ ìœ ì¼í•˜ì§€ ì•ŠìŒ)
                first_index[char] = -1
        
        # ìœ ì¼í•œ ë¬¸ìë“¤ ì¤‘ ê°€ì¥ ì‘ì€ ì¸ë±ìŠ¤ ì°¾ê¸°
        min_idx = float('inf')
        for char, idx in first_index.items():
            if idx != -1:
                min_idx = min(min_idx, idx)
        
        return min_idx if min_idx != float('inf') else -1


# ============================================================
# ì•Œê³ ë¦¬ì¦˜ ì‹œê°ì  ì„¤ëª…
# ============================================================
def visualize_algorithm(s: str):
    """
    ì•Œê³ ë¦¬ì¦˜ì˜ ë™ì‘ ê³¼ì •ì„ ì‹œê°ì ìœ¼ë¡œ ë³´ì—¬ì¤ë‹ˆë‹¤.
    """
    print(f"\n{'='*60}")
    print(f"ì…ë ¥ ë¬¸ìì—´: '{s}'")
    print(f"{'='*60}")
    
    if not s:
        print("ë¹ˆ ë¬¸ìì—´ â†’ ê²°ê³¼: -1")
        return
    
    # Step 1: ë¹ˆë„ ê³„ì‚° ì‹œê°í™”
    print("\n[Step 1] ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ ê³„ì‚°:")
    char_count = {}
    for i, char in enumerate(s):
        char_count[char] = char_count.get(char, 0) + 1
        print(f"  s[{i}] = '{char}' â†’ í˜„ì¬ ì¹´ìš´íŠ¸: {char_count}")
    
    print(f"\n  ìµœì¢… ë¹ˆë„ ë§µ: {char_count}")
    
    # Step 2: ìœ ì¼ ë¬¸ì íƒìƒ‰ ì‹œê°í™”
    print("\n[Step 2] ìœ ì¼ ë¬¸ì íƒìƒ‰:")
    for idx, char in enumerate(s):
        count = char_count[char]
        status = "âœ… ìœ ì¼!" if count == 1 else f"âŒ {count}ë²ˆ ë“±ì¥"
        print(f"  s[{idx}] = '{char}' â†’ ë“±ì¥ íšŸìˆ˜: {count} â†’ {status}")
        if count == 1:
            print(f"\n  ğŸ¯ ê²°ê³¼: ì¸ë±ìŠ¤ {idx} (ë¬¸ì '{char}')")
            return
    
    print("\n  ğŸ¯ ê²°ê³¼: -1 (ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µ)")


# ============================================================
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ ë° ìë™ ì±„ì 
# ============================================================

# (ì…ë ¥ ë¬¸ìì—´, ì˜ˆìƒ ì •ë‹µ)
test_cases = [
    ("morningstar", 0),
    ("lovealgorithm", 2),
    ("aabb", -1),
    ("abcdefg", 0),
    ("aabbccddeefg", 10),
    ("", -1),
    ("z", 0),
    ("aabbcc", -1),
    ("abcabcddxyz", 8),
]

# ìë™ ì±„ì 
sol = Solution()
score = 0

print("=" * 60)
print("ìë™ ì±„ì  ê²°ê³¼")
print("=" * 60)

for idx, (s, expected) in enumerate(test_cases, 1):
    result = sol.firstUniqChar(s)
    correct = (result == expected)
    print(f"Test {idx}: s='{s}' â†’ ê²°ê³¼={result}, ì˜ˆìƒ={expected} â†’ {'âœ… ì •ë‹µ' if correct else 'âŒ ì˜¤ë‹µ'}")
    if correct:
        score += 1

print(f"\nì´ ì ìˆ˜: {score}/{len(test_cases)}")

# ì•Œê³ ë¦¬ì¦˜ ì‹œê°í™” ì˜ˆì‹œ
visualize_algorithm("morningstar")
visualize_algorithm("aabb")
````
