````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  ë¬¸ì œ: Find First Unique Character (ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ì°¾ê¸°)                      â•‘
# â•‘  ì•Œê³ ë¦¬ì¦˜: HashMap (í•´ì‹œë§µ) í™œìš©                                                â•‘
# â•‘  ì‹œê°„ ë³µì¡ë„: O(n) - ë¬¸ìì—´ì„ ë‘ ë²ˆ ìˆœíšŒ                                         â•‘
# â•‘  ê³µê°„ ë³µì¡ë„: O(1) - ì˜ë¬¸ ì†Œë¬¸ì 26ê°œë¡œ ì œí•œ (ìƒìˆ˜)                               â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
#
# â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
# â”‚ ğŸ“Œ ë¬¸ì œë¥¼ í•œ ë¬¸ì¥ìœ¼ë¡œ ìš”ì•½                                                     â”‚
# â”‚ "ë¬¸ìì—´ì—ì„œ ë”± í•œ ë²ˆë§Œ ë“±ì¥í•˜ëŠ” ë¬¸ì ì¤‘ ê°€ì¥ ì•ì— ìˆëŠ” ê²ƒì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•˜ë¼"    â”‚
# â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
#
# â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
# â”‚ ğŸ“Œ í•µì‹¬ ì•„ì´ë””ì–´                                                              â”‚
# â”‚                                                                              â”‚
# â”‚ 1. í•´ì‹œë§µ(ë”•ì…”ë„ˆë¦¬)ì— ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ë¥¼ ì €ì¥                                â”‚
# â”‚ 2. ë¬¸ìì—´ì„ ì²˜ìŒë¶€í„° ë‹¤ì‹œ ìˆœíšŒí•˜ë©° ë“±ì¥ íšŸìˆ˜ê°€ 1ì¸ ì²« ë¬¸ì ì°¾ê¸°                  â”‚
# â”‚                                                                              â”‚
# â”‚ ì™œ ë‘ ë²ˆ ìˆœíšŒí•˜ëŠ”ê°€?                                                          â”‚
# â”‚ - ì²« ë²ˆì§¸ ìˆœíšŒ: ëª¨ë“  ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ë¥¼ ë¯¸ë¦¬ íŒŒì•…                               â”‚
# â”‚ - ë‘ ë²ˆì§¸ ìˆœíšŒ: "ì²« ë²ˆì§¸" ìœ ì¼ ë¬¸ìë¥¼ ì°¾ê¸° ìœ„í•´ ì•ì—ì„œë¶€í„° í™•ì¸                  â”‚
# â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
#
# â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
# â”‚ ğŸ“Œ Python í•µì‹¬ ë¬¸ë²• ì„¤ëª…                                                      â”‚
# â”‚                                                                              â”‚
# â”‚ 1. collections.Counter                                                       â”‚
# â”‚    - ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´ì˜ ê° ìš”ì†Œ ê°œìˆ˜ë¥¼ ìë™ìœ¼ë¡œ ì„¸ì–´ì¤Œ                          â”‚
# â”‚    - Counter("aab") â†’ {'a': 2, 'b': 1}                                       â”‚
# â”‚                                                                              â”‚
# â”‚ 2. enumerate(iterable)                                                       â”‚
# â”‚    - ì¸ë±ìŠ¤ì™€ ê°’ì„ ë™ì‹œì— ë°˜í™˜                                                 â”‚
# â”‚    - for idx, char in enumerate("abc"): â†’ (0,'a'), (1,'b'), (2,'c')          â”‚
# â”‚                                                                              â”‚
# â”‚ 3. dict.get(key, default)                                                    â”‚
# â”‚    - keyê°€ ì¡´ì¬í•˜ë©´ í•´ë‹¹ ê°’ ë°˜í™˜                                               â”‚
# â”‚    - keyê°€ ì—†ìœ¼ë©´ default ë°˜í™˜ (KeyError ë°©ì§€)                                â”‚
# â”‚    - ì˜ˆ: {}.get('a', 0) â†’ 0 (í‚¤ê°€ ì—†ìœ¼ë¯€ë¡œ ê¸°ë³¸ê°’ 0 ë°˜í™˜)                      â”‚
# â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

from collections import Counter
from typing import List

class Solution:
    """
    ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ì°¾ê¸° ë¬¸ì œ ì†”ë£¨ì…˜ í´ë˜ìŠ¤.

    First Unique Character problem solution class.

    ì´ í´ë˜ìŠ¤ëŠ” ì„¸ ê°€ì§€ ë°©ë²•ì„ ì œê³µí•©ë‹ˆë‹¤:
    1. Counter ì‚¬ìš© (ê°€ì¥ ê°„ê²°)
    2. ì¼ë°˜ dict ì‚¬ìš© (Counter ì—†ì´)
    3. ì¸ë±ìŠ¤ ì €ì¥ ë°©ì‹ (ìµœì í™”)
    """

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # ë°©ë²• 1: collections.Counter ì‚¬ìš© (ê°€ì¥ ê°„ê²°í•˜ê³  Pythonic)
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    def firstUniqChar(self, s: str) -> int:
        """
        ë¬¸ìì—´ì—ì„œ í•œ ë²ˆë§Œ ë“±ì¥í•˜ëŠ” ì²« ë²ˆì§¸ ë¬¸ìì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.

        Args:
            s: ì…ë ¥ ë¬¸ìì—´ (ì†Œë¬¸ì ì˜ì–´ ì•ŒíŒŒë²³ìœ¼ë¡œë§Œ êµ¬ì„±)

        Returns:
            int: ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ìì˜ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ -1

        Examples:
            >>> sol = Solution()
            >>> sol.firstUniqChar("morningstar")
            0  # 'm'ì´ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì (ì¸ë±ìŠ¤ 0)
            >>> sol.firstUniqChar("aabb")
            -1  # ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µ
        """

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ ì—£ì§€ ì¼€ì´ìŠ¤: ë¹ˆ ë¬¸ìì—´                                               â”‚
        # â”‚ - ë¹ˆ ë¬¸ìì—´ì—ëŠ” ìœ ì¼í•œ ë¬¸ìê°€ ì—†ìœ¼ë¯€ë¡œ -1 ë°˜í™˜                         â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        if not s:
            return -1

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ Step 1: Counterë¡œ ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ ê³„ì‚°                            â”‚
        # â”‚                                                                    â”‚
        # â”‚ CounterëŠ” ë‚´ë¶€ì ìœ¼ë¡œ ë”•ì…”ë„ˆë¦¬(í•´ì‹œë§µ)ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.                    â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì˜ˆ: s = "morningstar"                                               â”‚
        # â”‚     Counter(s) = {'m':1, 'o':1, 'r':2, 'n':2, 'i':1,               â”‚
        # â”‚                   'g':1, 's':1, 't':1, 'a':1}                       â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì‹œê°„: O(n) - ë¬¸ìì—´ì„ í•œ ë²ˆ ìˆœíšŒ                                     â”‚
        # â”‚ ê³µê°„: O(1) - ìµœëŒ€ 26ê°œì˜ ì˜ë¬¸ ì†Œë¬¸ì (ìƒìˆ˜)                           â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        char_count = Counter(s)

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ Step 2: ë¬¸ìì—´ì„ ìˆœíšŒí•˜ë©° ë“±ì¥ íšŸìˆ˜ê°€ 1ì¸ ì²« ë²ˆì§¸ ë¬¸ì ì°¾ê¸°            â”‚
        # â”‚                                                                    â”‚
        # â”‚ enumerate(s): ì¸ë±ìŠ¤(idx)ì™€ ë¬¸ì(char)ë¥¼ ë™ì‹œì— ë°˜í™˜                  â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì™œ Counterê°€ ì•„ë‹Œ ì›ë³¸ ë¬¸ìì—´ì„ ìˆœíšŒí•˜ëŠ”ê°€?                           â”‚
        # â”‚ - Counterì˜ ìˆœì„œëŠ” ì‚½ì… ìˆœì„œë¥¼ ë”°ë¥´ì§€ë§Œ (Python 3.7+)                â”‚
        # â”‚ - ìš°ë¦¬ëŠ” "ì›ë³¸ ë¬¸ìì—´ì—ì„œì˜ ìœ„ì¹˜(ì¸ë±ìŠ¤)"ê°€ í•„ìš”í•˜ë¯€ë¡œ                 â”‚
        # â”‚ - ì›ë³¸ ë¬¸ìì—´ì„ ìˆœíšŒí•´ì•¼ ì •í™•í•œ ì¸ë±ìŠ¤ë¥¼ ì–»ì„ ìˆ˜ ìˆìŒ                  â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì‹œê°„: O(n) - ìµœì•…ì˜ ê²½ìš° ë¬¸ìì—´ ëê¹Œì§€ ìˆœíšŒ                           â”‚
        # â”‚       (ìµœì„ ì˜ ê²½ìš° ì²« ë¬¸ìê°€ ìœ ì¼í•˜ë©´ O(1))                           â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        for idx, char in enumerate(s):
            # char_count[char]: O(1) ì‹œê°„ì— í•´ë‹¹ ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ ì¡°íšŒ
            if char_count[char] == 1:
                return idx  # ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ë°œê²¬! ì¦‰ì‹œ ë°˜í™˜

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µì¸ ê²½ìš°                                              â”‚
        # â”‚ - ë£¨í”„ê°€ ëê¹Œì§€ ëŒì•˜ë‹¤ëŠ” ê²ƒì€ ë“±ì¥ íšŸìˆ˜ê°€ 1ì¸ ë¬¸ìê°€ ì—†ë‹¤ëŠ” ì˜ë¯¸       â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        return -1


    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # ë°©ë²• 2: ì¼ë°˜ ë”•ì…”ë„ˆë¦¬ ì‚¬ìš© (Counter ì—†ì´ ì§ì ‘ êµ¬í˜„)
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    def firstUniqChar_manual(self, s: str) -> int:
        """
        ë°©ë²• 2: Counter ì—†ì´ ì¼ë°˜ ë”•ì…”ë„ˆë¦¬ë¡œ ì§ì ‘ êµ¬í˜„.

        Counter ì‚¬ìš©ì´ ì œí•œëœ ë©´ì ‘ ìƒí™©ì—ì„œ ìœ ìš©í•©ë‹ˆë‹¤.
        í•´ì‹œë§µì˜ ë™ì‘ ì›ë¦¬ë¥¼ ì´í•´í•˜ê¸° ìœ„í•œ í•™ìŠµìš©ìœ¼ë¡œë„ ì¢‹ìŠµë‹ˆë‹¤.

        Args:
            s: ì…ë ¥ ë¬¸ìì—´

        Returns:
            int: ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ìì˜ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ -1
        """

        if not s:
            return -1

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ Step 1: ë¹ˆ í•´ì‹œë§µ(ë”•ì…”ë„ˆë¦¬) ìƒì„±                                      â”‚
        # â”‚                                                                    â”‚
        # â”‚ ë”•ì…”ë„ˆë¦¬ì˜ êµ¬ì¡°:                                                    â”‚
        # â”‚   key: ë¬¸ì (ì˜ˆ: 'a', 'b', 'c')                                     â”‚
        # â”‚   value: í•´ë‹¹ ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ (ì˜ˆ: 1, 2, 3)                         â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        char_count = {}

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ Step 2: ì²« ë²ˆì§¸ ìˆœíšŒ - ê° ë¬¸ìì˜ ë¹ˆë„ ê³„ì‚°                            â”‚
        # â”‚                                                                    â”‚
        # â”‚ dict.get(key, default)ì˜ ë™ì‘:                                      â”‚
        # â”‚   - keyê°€ ë”•ì…”ë„ˆë¦¬ì— ìˆìœ¼ë©´: í•´ë‹¹ value ë°˜í™˜                         â”‚
        # â”‚   - keyê°€ ë”•ì…”ë„ˆë¦¬ì— ì—†ìœ¼ë©´: default ê°’ ë°˜í™˜ (ì—¬ê¸°ì„œëŠ” 0)             â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì˜ˆì‹œ (s = "aab"):                                                   â”‚
        # â”‚   char='a': char_count.get('a', 0) = 0 â†’ char_count['a'] = 1       â”‚
        # â”‚   char='a': char_count.get('a', 0) = 1 â†’ char_count['a'] = 2       â”‚
        # â”‚   char='b': char_count.get('b', 0) = 0 â†’ char_count['b'] = 1       â”‚
        # â”‚   ê²°ê³¼: {'a': 2, 'b': 1}                                            â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        for char in s:
            # í˜„ì¬ ì¹´ìš´íŠ¸ë¥¼ ê°€ì ¸ì˜¤ê³  (ì—†ìœ¼ë©´ 0), 1ì„ ë”í•´ì„œ ì €ì¥
            char_count[char] = char_count.get(char, 0) + 1

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ Step 3: ë‘ ë²ˆì§¸ ìˆœíšŒ - ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì ì°¾ê¸°                         â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì›ë³¸ ë¬¸ìì—´ì˜ ìˆœì„œëŒ€ë¡œ í™•ì¸í•´ì•¼ "ì²« ë²ˆì§¸" ìœ ì¼ ë¬¸ìë¥¼ ì°¾ì„ ìˆ˜ ìˆìŒ      â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        for idx, char in enumerate(s):
            if char_count[char] == 1:
                return idx

        return -1


    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # ë°©ë²• 3: ì¸ë±ìŠ¤ ì €ì¥ ë°©ì‹ (í•œ ë²ˆì˜ ìˆœíšŒë¡œ í•´ê²° ì‹œë„)
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    def firstUniqChar_optimized(self, s: str) -> int:
        """
        ë°©ë²• 3: ê° ë¬¸ìì˜ ì²« ë“±ì¥ ì¸ë±ìŠ¤ë¥¼ ì €ì¥í•˜ê³ , ì¤‘ë³µ ì‹œ ë§ˆí‚¹.

        ì•„ì´ë””ì–´:
        - ê° ë¬¸ìê°€ ì²˜ìŒ ë“±ì¥í•œ ìœ„ì¹˜(ì¸ë±ìŠ¤)ë¥¼ ì €ì¥
        - ê°™ì€ ë¬¸ìê°€ ë‹¤ì‹œ ë“±ì¥í•˜ë©´ -1ë¡œ ë§ˆí‚¹ (ë” ì´ìƒ ìœ ì¼í•˜ì§€ ì•ŠìŒ)
        - ë§ˆì§€ë§‰ì— -1ì´ ì•„ë‹Œ ê²ƒ ì¤‘ ê°€ì¥ ì‘ì€ ì¸ë±ìŠ¤ ë°˜í™˜

        ì¥ì : ë¬¸ìì—´ì´ ë§¤ìš° ê¸¸ê³  ìœ ì¼ ë¬¸ìê°€ ì•ìª½ì— ìˆì–´ë„ ìµœì í™” ê°€ëŠ¥
        ë‹¨ì : êµ¬í˜„ì´ ë³µì¡í•˜ê³ , ìµœì•…ì˜ ê²½ìš° ì„±ëŠ¥ ì°¨ì´ ì—†ìŒ

        Args:
            s: ì…ë ¥ ë¬¸ìì—´

        Returns:
            int: ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ìì˜ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ -1
        """

        if not s:
            return -1

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ í•´ì‹œë§µ êµ¬ì¡°:                                                        â”‚
        # â”‚   key: ë¬¸ì                                                        â”‚
        # â”‚   value: ì²« ë“±ì¥ ì¸ë±ìŠ¤ (ì¤‘ë³µ ë°œìƒ ì‹œ -1ë¡œ ë³€ê²½)                      â”‚
        # â”‚                                                                    â”‚
        # â”‚ ì˜ˆì‹œ (s = "aba"):                                                   â”‚
        # â”‚   i=0, 'a': first_index = {'a': 0}                                 â”‚
        # â”‚   i=1, 'b': first_index = {'a': 0, 'b': 1}                         â”‚
        # â”‚   i=2, 'a': first_index = {'a': -1, 'b': 1}  â† ì¤‘ë³µì´ë¯€ë¡œ -1        â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        first_index = {}

        for idx, char in enumerate(s):
            if char not in first_index:
                # ì²˜ìŒ ë“±ì¥: í•´ë‹¹ ì¸ë±ìŠ¤ ì €ì¥
                first_index[char] = idx
            else:
                # ì¤‘ë³µ ë°œìƒ: -1ë¡œ ë§ˆí‚¹ (ì´ ë¬¸ìëŠ” ë” ì´ìƒ ìœ ì¼í•˜ì§€ ì•ŠìŒ)
                first_index[char] = -1

        # â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        # â”‚ ìœ ì¼í•œ ë¬¸ìë“¤ ì¤‘ ê°€ì¥ ì‘ì€ ì¸ë±ìŠ¤ ì°¾ê¸°                                â”‚
        # â”‚                                                                    â”‚
        # â”‚ first_index.items(): (key, value) ìŒì„ ë°˜ë³µ                         â”‚
        # â”‚ idx != -1: ìœ ì¼í•œ ë¬¸ìë§Œ ê³ ë ¤ (ì¤‘ë³µëœ ë¬¸ìëŠ” -1ì´ë¯€ë¡œ ì œì™¸)           â”‚
        # â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
        min_idx = float('inf')  # ë¬´í•œëŒ€ë¡œ ì´ˆê¸°í™” (ì–´ë–¤ ì¸ë±ìŠ¤ë³´ë‹¤ë„ í¼)

        for char, idx in first_index.items():
            if idx != -1:  # ìœ ì¼í•œ ë¬¸ìì¸ ê²½ìš°ë§Œ
                min_idx = min(min_idx, idx)  # ë” ì‘ì€ ì¸ë±ìŠ¤ë¡œ ê°±ì‹ 

        # ìœ ì¼í•œ ë¬¸ìê°€ ìˆìœ¼ë©´ í•´ë‹¹ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ -1 ë°˜í™˜
        return min_idx if min_idx != float('inf') else -1


# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  ğŸ“Š ì•Œê³ ë¦¬ì¦˜ ì‹œê°ì  ì„¤ëª… í•¨ìˆ˜                                                   â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
def visualize_algorithm(s: str):
    """
    ì•Œê³ ë¦¬ì¦˜ì˜ ë™ì‘ ê³¼ì •ì„ ì‹œê°ì ìœ¼ë¡œ ë³´ì—¬ì¤ë‹ˆë‹¤.

    ê° ë‹¨ê³„ì—ì„œ í•´ì‹œë§µì´ ì–´ë–»ê²Œ ë³€í™”í•˜ëŠ”ì§€,
    ìµœì¢…ì ìœ¼ë¡œ ì–´ë–¤ ë¬¸ìê°€ ì„ íƒë˜ëŠ”ì§€ ë³´ì—¬ì¤ë‹ˆë‹¤.

    Args:
        s: ë¶„ì„í•  ë¬¸ìì—´
    """
    print(f"\n{'='*60}")
    print(f"ì…ë ¥ ë¬¸ìì—´: '{s}'")
    print(f"ë¬¸ìì—´ ê¸¸ì´: {len(s)}")
    print(f"{'='*60}")

    if not s:
        print("ë¹ˆ ë¬¸ìì—´ â†’ ê²°ê³¼: -1")
        return

    # Step 1: ë¹ˆë„ ê³„ì‚° ì‹œê°í™”
    print("\n[Step 1] ê° ë¬¸ìì˜ ë“±ì¥ íšŸìˆ˜ ê³„ì‚° (ì²« ë²ˆì§¸ ìˆœíšŒ)")
    print("-" * 60)

    char_count = {}
    for i, char in enumerate(s):
        old_count = char_count.get(char, 0)
        char_count[char] = old_count + 1
        print(f"  s[{i:2d}] = '{char}' â†’ "
              f"count['{char}']: {old_count} â†’ {char_count[char]} | "
              f"í˜„ì¬ ë§µ: {dict(sorted(char_count.items()))}")

    print(f"\n  âœ… ìµœì¢… ë¹ˆë„ ë§µ: {dict(sorted(char_count.items()))}")

    # ë¹ˆë„ ë¶„ì„
    unique_chars = [c for c, cnt in char_count.items() if cnt == 1]
    duplicate_chars = [c for c, cnt in char_count.items() if cnt > 1]
    print(f"  ğŸ“Š ìœ ì¼ ë¬¸ì (count=1): {unique_chars if unique_chars else 'ì—†ìŒ'}")
    print(f"  ğŸ“Š ì¤‘ë³µ ë¬¸ì (count>1): {duplicate_chars if duplicate_chars else 'ì—†ìŒ'}")

    # Step 2: ìœ ì¼ ë¬¸ì íƒìƒ‰ ì‹œê°í™”
    print("\n[Step 2] ìœ ì¼ ë¬¸ì íƒìƒ‰ (ë‘ ë²ˆì§¸ ìˆœíšŒ)")
    print("-" * 60)

    for idx, char in enumerate(s):
        count = char_count[char]
        if count == 1:
            print(f"  s[{idx:2d}] = '{char}' â†’ ë“±ì¥ íšŸìˆ˜: {count} â†’ âœ… ìœ ì¼! íƒìƒ‰ ì¢…ë£Œ")
            print(f"\n{'='*60}")
            print(f"ğŸ¯ ê²°ê³¼: ì¸ë±ìŠ¤ {idx} (ë¬¸ì '{char}')")
            print(f"{'='*60}")
            return
        else:
            print(f"  s[{idx:2d}] = '{char}' â†’ ë“±ì¥ íšŸìˆ˜: {count} â†’ âŒ ì¤‘ë³µ ({count}ë²ˆ ë“±ì¥)")

    print(f"\n{'='*60}")
    print("ğŸ¯ ê²°ê³¼: -1 (ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µ)")
    print(f"{'='*60}")


# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  ğŸ” ë¸Œë£¨íŠ¸í¬ìŠ¤ vs í•´ì‹œë§µ ì„±ëŠ¥ ë¹„êµ                                              â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
def compare_approaches(s: str):
    """
    ë¸Œë£¨íŠ¸í¬ìŠ¤ ë°©ì‹ê³¼ í•´ì‹œë§µ ë°©ì‹ì˜ ì°¨ì´ë¥¼ ì‹œê°ì ìœ¼ë¡œ ë³´ì—¬ì¤ë‹ˆë‹¤.

    ì‹¤ì œ ì—°ì‚° íšŸìˆ˜ë¥¼ ì„¸ì–´ O(nÂ²)ê³¼ O(n)ì˜ ì°¨ì´ë¥¼ ì²´ê°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    """
    print(f"\n{'='*60}")
    print(f"ë¸Œë£¨íŠ¸í¬ìŠ¤ vs í•´ì‹œë§µ ë¹„êµ")
    print(f"ì…ë ¥: '{s}' (ê¸¸ì´: {len(s)})")
    print(f"{'='*60}")

    # ë¸Œë£¨íŠ¸í¬ìŠ¤ ì—°ì‚° íšŸìˆ˜ ê³„ì‚°
    n = len(s)
    bruteforce_ops = n * n  # O(nÂ²)
    hashmap_ops = n + n     # O(n) + O(n)

    print(f"\n[ë¸Œë£¨íŠ¸í¬ìŠ¤ O(nÂ²)]")
    print(f"  - ê° ë¬¸ìë§ˆë‹¤ ì „ì²´ ë¬¸ìì—´ ìˆœíšŒ")
    print(f"  - ì˜ˆìƒ ì—°ì‚° íšŸìˆ˜: {n} Ã— {n} = {bruteforce_ops:,}")

    print(f"\n[í•´ì‹œë§µ O(n)]")
    print(f"  - ì²« ë²ˆì§¸ ìˆœíšŒ: ë¹ˆë„ ê³„ì‚° ({n}íšŒ)")
    print(f"  - ë‘ ë²ˆì§¸ ìˆœíšŒ: ìœ ì¼ ë¬¸ì íƒìƒ‰ (ìµœëŒ€ {n}íšŒ)")
    print(f"  - ì˜ˆìƒ ì—°ì‚° íšŸìˆ˜: {n} + {n} = {hashmap_ops:,}")

    print(f"\n[ì„±ëŠ¥ ì°¨ì´]")
    if bruteforce_ops > 0:
        ratio = bruteforce_ops / hashmap_ops
        print(f"  - ë¸Œë£¨íŠ¸í¬ìŠ¤/í•´ì‹œë§µ = {ratio:.1f}ë°° ì°¨ì´")
        print(f"  - ë¬¸ìì—´ì´ ê¸¸ìˆ˜ë¡ ì°¨ì´ê°€ ì»¤ì§!")


# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  ğŸ§ª í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ ë° ìë™ ì±„ì                                                   â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# (ì…ë ¥ ë¬¸ìì—´, ì˜ˆìƒ ì •ë‹µ, ì„¤ëª…)
test_cases = [
    ("morningstar", 0, "'m'ì´ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì"),
    ("lovealgorithm", 2, "'v'ê°€ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì (l,oëŠ” ì¤‘ë³µ)"),
    ("aabb", -1, "ëª¨ë“  ë¬¸ìê°€ ì¤‘ë³µ"),
    ("abcdefg", 0, "ëª¨ë“  ë¬¸ìê°€ ìœ ì¼, ì²« ë²ˆì§¸ê°€ ì •ë‹µ"),
    ("aabbccddeefg", 10, "'f'ê°€ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì"),
    ("", -1, "ë¹ˆ ë¬¸ìì—´"),
    ("z", 0, "ë‹¨ì¼ ë¬¸ì"),
    ("aabbcc", -1, "ëª¨ë“  ë¬¸ìê°€ ì •í™•íˆ 2ë²ˆì”©"),
    ("abcabcddxyz", 8, "'x'ê°€ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì"),
    ("leetcode", 0, "'l'ì´ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì"),
    ("loveleetcode", 2, "'v'ê°€ ì²« ë²ˆì§¸ ìœ ì¼ ë¬¸ì"),
]

# ìë™ ì±„ì 
sol = Solution()
score = 0

print("=" * 70)
print("ìë™ ì±„ì  ê²°ê³¼")
print("=" * 70)

for idx, (s, expected, desc) in enumerate(test_cases, 1):
    result = sol.firstUniqChar(s)
    correct = (result == expected)
    status = 'âœ…' if correct else 'âŒ'
    print(f"Test {idx:2d}: s='{s}' â†’ ê²°ê³¼={result}, ì˜ˆìƒ={expected} {status}")
    print(f"         ì„¤ëª…: {desc}")
    if correct:
        score += 1

print(f"\n{'='*70}")
print(f"ì´ ì ìˆ˜: {score}/{len(test_cases)}")
print(f"{'='*70}")

# ì•Œê³ ë¦¬ì¦˜ ì‹œê°í™” ì˜ˆì‹œ
print("\n\n" + "=" * 70)
print("ğŸ“Š ì•Œê³ ë¦¬ì¦˜ ë™ì‘ ê³¼ì • ì‹œê°í™”")
print("=" * 70)

visualize_algorithm("morningstar")
visualize_algorithm("aabb")
visualize_algorithm("leetcode")

# ì„±ëŠ¥ ë¹„êµ
compare_approaches("morningstar")
compare_approaches("a" * 1000)  # ê¸´ ë¬¸ìì—´ë¡œ ì°¨ì´ í™•ì¸
````
